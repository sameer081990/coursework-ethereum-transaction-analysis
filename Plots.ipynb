```python
df = pd.read_json('scams.json', orient='records')
```

```python
df2 = pd.json_normalize(df['result'])
```

```python
df2
```

```python
df2['subcategory'] = df2.apply(lambda row: str(row['subcategory']).capitalize(), axis=1)
```

```python
df2
```

```python
addr_dict = {}
def get_addresses(row):
    if(row['coin'] == 'ETH'):
        if (isinstance(row['addresses'], list)):
            for addr in row['addresses']:
                if (addr[0:2] == '0x'):
                    if(not addr in addr_dict): addr_dict[addr] = row['subcategory']

        if (isinstance(row['addresses'], str)):
            if (row['addresses'][0:2] == '0x'):
                if(not row['addresses'] in addr_dict): addr_dict[row['addresses']] = row['subcategory']

df2.apply(lambda x: get_addresses(x), axis=1)
```

```python
gas_supplied = pd.read_csv('weekly_gas_supplied.csv')
```

```python
#plt.figure(figsize=(18,18))
fig, ax = plt.subplots(nrows=6, ncols=1)
fig.set_figheight(29.7/1.5)
fig.set_figwidth(21/1.5)

plot_all = sb.lineplot(data=gas_supplied, x='YearMonthWeek', y='AverageGas', hue='Type', ax=ax[0])
plot_all.set_xlabel('')
plot_all.set_title('Average weekly gas supply in transactions - Since launch to 2019')

plot_2015 = sb.lineplot(data=gas_supplied[gas_supplied['Year'] == 2015], x='WeekNum', y='AverageGas', hue='Type', ax=ax[1])
plot_2015.set_xlabel('')
plot_2015.set_title('Average weekly gas supply in transactions  - 2015')

plot_2016 = sb.lineplot(data=gas_supplied[gas_supplied['Year'] == 2016], x='WeekNum', y='AverageGas', hue='Type', ax=ax[2])
plot_2016.set_xlabel('')
plot_2016.set_title('Average weekly gas supply in transactions - 2016')

plot_2017 = sb.lineplot(data=gas_supplied[gas_supplied['Year'] == 2017], x='WeekNum', y='AverageGas', hue='Type', ax=ax[3])
plot_2017.set_xlabel('')
plot_2017.set_title('Average weekly gas supply in transactions - 2017')

plot_2018 = sb.lineplot(data=gas_supplied[gas_supplied['Year'] == 2018], x='WeekNum', y='AverageGas', hue='Type', ax=ax[4])
plot_2018.set_xlabel('')
plot_2018.set_title('Average weekly gas supply in transactions - 2018')

plot_2019 = sb.lineplot(data=gas_supplied[gas_supplied['Year'] == 2019], x='WeekNum', y='AverageGas', hue='Type', ax=ax[5])
plot_2019.set_xlabel('Week number')
plot_2019.set_title('Average weekly gas supply in transactions - 2019')

plt.show()
```
